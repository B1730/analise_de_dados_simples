```python
#Conjunto de dados de itens do pedido, onde precisei retirar as aspas duplas e converter a coluna 'shipping_limit_date' para o formato datetime.
conjunto_dados_itens_pedido = pd.read_csv("../datasets/dataset/dataset_original/olist_order_items_dataset.csv")
conjunto_dados_itens_pedido.columns = conjunto_dados_itens_pedido.columns.str.strip()
conjunto_dados_itens_pedido.columns = conjunto_dados_itens_pedido.columns.str.replace('"',"")
conjunto_dados_itens_pedido['shipping_limit_date']
conjunto_dados_itens_pedido.isnull().sum()
conjunto_dados_itens_pedido[['product_id','seller_id']] =(
    conjunto_dados_itens_pedido[['product_id','seller_id']].apply(lambda col: col.str.replace('"', ''))
)
conjunto_dados_itens_pedido.to_csv("../datasets/dataset/dataset_modificado/olist_order_items_dataset_modificado.csv", index=False)
conjunto_dados_itens_pedido_modificado = pd.read_csv("../datasets/dataset/dataset_modificado/olist_order_items_dataset_modificado.csv")
conjunto_dados_itens_pedido_modificado['shipping_limit_date'] = pd.to_datetime(conjunto_dados_itens_pedido['shipping_limit_date'], errors='coerce')
conjunto_dados_itens_pedido_modificado.info()
conjunto_dados_itens_pedido_modificado
```
